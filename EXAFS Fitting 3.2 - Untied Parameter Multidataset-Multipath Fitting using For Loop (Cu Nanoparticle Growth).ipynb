```python
"""
--- Multi-spectra EXAFS Fitting w/ multiple scatter path ---


This example demostrates EXAFS fitting of a time series data of Cu nanoparticle growth from Cu(Ac)2 precursor.
Multiple spectra and scattering paths (Cu-Cu and Cu-O) are fitted.
A custom module featuring parameter table is used.
For loop is used to fit each spectra independently.

"""
```

```python
# Base packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', None)

import os
import h5py

# Packages for reading data and normalisation
from larch.io import read_ascii
from larch.xafs import autobk, xftf

from tools import exafs_param_table as pt

# Packages for EXAFS fittings
from larch.xafs import feffit_transform, feffit_dataset, feffit

# Packages for reading EXAFS fittings results
from larch.xafs import feffit_report

from tools import io_results
```

```python
""" 0. Preparing Data. As an example a time series of 10 spectra are loaded. """

data_dir = r"data\exafs_fitting\time_series"
fnames = [fname for fname in os.listdir(data_dir)]
path_list = [os.path.join(data_dir, fname) for fname in fnames]
```

```python
""" 1. Batch Reading Data """

# Initialisation
data_list = [read_ascii(fpath, labels="null energy mu") for fpath in path_list]
for data, fname in zip(data_list, fnames):
    data.groupname = fname

# Pre-processing - EXAFS fittings only need a valid k space
    autobk(data, pre_edge_kws={"nnorm": 0, "nvict": 0}, rbkg=1.0)
    xftf(data, kmin=3.0, kmax=10., dk=0.5, window="hanning", kweight=2.00)
```

```python
""" Misc: Plotting a spectrum from the time series as inspection. """
fig, ax = plt.subplots(3, 1, figsize=(12, 24))

count = 0
for data in data_list:
    ax[0].plot(data.energy, data.norm + (count * 0.3))
    ax[1].plot(data.k, data.chi * data.k ** 2 + (count * 0.5))
    ax[2].plot(data.r, data.chir_mag + (count * 0.2))
    count += 1
    
ax[0].axhline(1, color="r", linestyle=":")

ax[0].set_xlabel("Energy (eV)")
ax[0].set_ylabel("Normalised Absorption")

ax[1].set_xlabel("k ($\AA^{-1}$)")
ax[1].set_ylabel("$\chi$(k) k$^2$")

ax[2].set_xlabel("R ($\AA$)")
ax[2].set_ylabel("$\chi$ R")
```

```python
""" 
Scattering paths generated from feff calculations 

In this case, the time series spectra are early stages of Cu nanoparticles formation from Cu(Ac)2 precursor measured in I20 EDE.

Previously computed Copper and Copper(II) oxide scattering paths are used for EXAFS fittings.
(See "Demonstration - Feff Calculation.ipynb")

"""
feff_folder_cuo =  r"data/feffs/Cu1_K_c7_Tenorite_cif11639"
feff_folder_cu =  r"data/feffs/Cu1_K_c7_Copper_cif11145"
```

```python
"""
A parameters table with all available scattering paths can be generated for fittings.

Notes: Param tables are stored as pandas dataframes, which means
    - Param table can be manipulated as any pandas dataframe.
    - Param table can be import/export as .csv files.

"""

fit_table_cu = pt.param_table(feff_folder_cu)
fit_table_cu
```

```python
""" 
With some manipulation with the dataframe. 

1. Make a param table for both Cu and CuO.
2. Pick out only the first scatter path.
3. Combine both scatter paths.

"""
# First scatter path of Cu
fit_table_cu = pt.param_table(feff_folder_cu)
fit_table_cu = fit_table_cu[fit_table_cu["reff"] < 3]

# First scatter path of CuO
fit_table_cuo = pt.param_table(feff_folder_cuo)
fit_table_cuo = fit_table_cuo[fit_table_cuo["reff"] < 1.9475]

# Combine both scatter paths in one paramter table
param_table = pd.concat([fit_table_cu, fit_table_cuo])
param_table
```

```python
"""
A series of different fitting constraints

1. constant s02 of 0.918 (obtained from fitting standard spectra.)
2. set degen as a fitting parameter instead of constant
3. group parameters based on different scattering path

Note the unique parameters the table.

Only these parameters are read into larch fitting
    - s02
    - e0
    - degen
    - sigma2
    - deltar
    - reff (read as parameter, however not in EXAFS equation)
    - fpath

"""
fit_table = param_table

# 1. Set s02 as constant for the fit
fit_table["s02"] = 0.918

# 2. Set degeneracy as a fitting parameter
fit_table["degen"] = "n"

# 3. Group parameters based on scattering path (Note the subscripts added based on scattering material)
fit_table = pt.groupby_path(fit_table, ["degen", "sigma2", "deltar"]) 

fit_table
```

```python
""" Export/save the fit table. """
output_dir = r"data/output/cuNPgrowth_loop/"
pt.save_table(fit_table, output_dir)
```

```python
""" Formulate the feffit parameters and paths using the parameters table (custom module). """

# Setup fitting parameters and scattering paths
feffit_params = pt.setup_feffit_params(fit_table)
feffit_paths = pt.setup_feffit_paths(fit_table)

# Define fit space and forward FT (larch function) (Note: rbkg needs to be set during autobk.)
feffit_trans = feffit_transform(kmin=3.0, kmax=10.000, dk=0.500, kw=2,
                                 window="hanning", fitspace='r', rmin=1.000, rmax=3.000)
```

```python
""" For Loop Export/save the fit table. """
for data in data_list:
    
    # Create a subfolder to contain each fit result
    output_subfolder = os.path.join(output_dir, data.groupname[-8:-4]) # use 
    os.mkdir(output_subfolder)

    # Construct fitting dataset
    feffit_datasets = feffit_dataset(data=data, transform=feffit_trans, paths=feffit_paths)
    
    # Final fitting and saving
    feffit_result = feffit(feffit_params, feffit_datasets)    
    io_results.save(feffit_result, output_subfolder)
```

```python
""" Summarise fitted parameters from each result. """

headers = ["file_index",
           "chi2_reduced", 
           "rfactor", 
           "n_Cu", 
           "n_CuO", 
           "sigma2_Cu", 
           "sigma2_CuO",
           "e0",
           "deltar_Cu", 
           "deltar_CuO"]
results = []

df = pd.DataFrame(columns=headers)

for file in os.listdir(output_dir):
    fpath = os.path.join(output_dir, file)
    if os.path.isdir(fpath):
        pkl_path = os.path.join(fpath, "feffit_result.pkl")
        result = io_results.load(pkl_path)
        results.append(result)

        dataroll = pd.Series([int(file),
                              result.chi2_reduced, 
                              result.rfactor, 
                              result.paramgroup.n_Cu._val, 
                              result.paramgroup.n_CuO._val,
                              result.paramgroup.sigma2_Cu._val, 
                              result.paramgroup.sigma2_CuO._val, 
                              result.paramgroup.e0._val, 
                              result.paramgroup.deltar_Cu._val, 
                              result.paramgroup.deltar_CuO._val], index = headers)
        df = df.append(dataroll, ignore_index=True)

# Save .csv file
df.to_csv(os.path.join(output_dir, "result_summary.csv"))
df
```

```python
""" Plotting Fitted Results. """
for result in results:

    fig, ax = plt.subplots(1, 2, figsize=(20, 7))

    spectra = result.datasets[0].data
    fit = result.datasets[0].model

    ax[0].plot(fit.r, fit.chir_mag, color="r", label="fitting")
    ax[0].plot(spectra.r, spectra.chir_mag, lw=0, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
    ax[0].axvline(1)
    ax[0].axvline(3)
    ax[0].set_xlim(0, 6)
    ax[0].legend()

    ax[1].plot(fit.k, fit.chi * fit.k ** 2, color="r", label="fitting")
    ax[1].plot(spectra.k, spectra.chi * spectra.k ** 2, lw=1, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
    ax[1].plot(fit.k, fit.kwin)
    ax[1].legend()
```
