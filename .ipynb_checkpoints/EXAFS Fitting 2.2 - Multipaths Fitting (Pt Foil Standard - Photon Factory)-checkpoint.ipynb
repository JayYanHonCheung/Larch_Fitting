```python
"""
--- Pt Standard EXAFS Fitting w/ Multiple Scattering Paths --- 

This example demostrates EXAFS fitting of a Pt foil standard with multiple scattering path using xraylarch.
Pt foil was measured in Photon Factory.
A custom module featuring parameter table is used.

"""
```

```python
# Base packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_rows', None)

import os

# Packages for reading data and normalisation
from larch.xafs import autobk, xftf, xftr

from tools import io_photon_factory as io_pf
from tools import exafs_param_table as pt

# Packages for EXAFS fittings
from larch.xafs import feffit_transform, feffit_dataset, feffit

# Packages for reading EXAFS fittings results
from larch.xafs import feffit_report

from tools import io_results 
```

```python
""" Loading and Pre-process Pt Foil Standard Spectrum """
fpath = r"data\standards\PhotonFactory_PtFoil_2_0.qd"
data = io_pf.read_trans(fpath, name="Photon Factory Pt Foil") # Custom function to load PF transmission data
autobk(data, rbkg=1.0) # auto-background substraction
xftf(data, kmin=2.8, kmax=14.4, dk=0.5, window="hanning", kweight=2.00) # Forward FT
xftr(data, qmax_out=15, rmin=1.0, rmax=6.5, dk=0.5, window="hanning") # Backwards FT

""" Plotting """
fig, ax = plt.subplots(2, 2, figsize=(15, 11))

ax[0, 0].plot(data.energy, data.mu, label="Fluorescence Signal", color="k")
ax[0, 0].plot(data.energy, data.pre_edge, label="pre-edge line")
ax[0, 0].plot(data.energy, data.post_edge, label="post-edge line")
ax[0, 0].set_xlabel("Energy (eV)")
ax[0, 0].set_ylabel("Signal (a.u.)")

ax[0, 1].plot(data.energy, data.norm, label="Normalised Absorption", color="k")
ax[0, 1].axhline(1, linestyle=":", color="r")
ax[0, 1].set_xlabel("Energy (eV)")
ax[0, 1].set_ylabel("Normalisation Absorption (a.u.)")

ax[1, 0].plot(data.k, data.chi * data.k ** 2, label="k-space", color="k")
ax[1, 0].plot(data.q, data.chiq, label="q-space", color="b")
ax[1, 0].plot(data.k, data.kwin, label="hanning window", color="r")
ax[1, 0].set_xlabel("k or q-space ($\AA^{-1}$)")
ax[1, 0].set_ylabel("$\chi$(k) k$^2$ or $\chi$(q) q$^2$")

ax[1, 1].plot(data.r, data.chir_mag, label="R-space", color="k")
ax[1, 1].plot(data.r, data.rwin, label="hanning window", color="r")
ax[1, 1].set_xlabel("R ($\AA$)")
ax[1, 1].set_ylabel("$\chi$ R")

for i in range(2):
    for j in range(2):
        ax[i, j].legend()
```

```python
# Fit table
feff_folder = r"data/feffs/Pt1_L3_c8_Pt_1011103"
feff_table = pt.param_table(feff_folder)
feff_table
```

```python
# Construct fit table (See more detail in "EXAFS Fitting 2.1")

# 1. Fit paths with effective R < 6.5
fit_table = feff_table[feff_table["reff"] < 6.5]

# 2. Assume paths with same reff belongs to the same shell
fit_table = pt.form_shells(fit_table)

# 3. Group sigma2 based on shell
fit_table = pt.groupby_shell(fit_table, ["sigma2"])

# 4. Separate e0 into 1st shell and the rest
fit_table = pt._1st_shell_e0(fit_table)

# 5. Introduce expansion/contraction factor (alpha)
fit_table["deltar"] = "reff*alpha"
fit_table.insert(11, "alpha", "alpha") # Add a new column for alpha

fit_table
```

```python
""" Formulate the feffit parameters and paths using the parameters table (custom module). """

# Setup
feffit_params = pt.setup_feffit_params(fit_table)
feffit_paths = pt.setup_feffit_paths(fit_table)

# Define fit space and forward FT (larch function)
feffit_trans = feffit_transform(kmin=2.8, kmax=14.400, dk=0.500, kw=2,
                                 window="hanning", fitspace='r', rmin=1.000, rmax=6.500)

# Construct fitting dataset 
feffit_datasets = feffit_dataset(data=data, transform=feffit_trans, paths=feffit_paths)
```

```python
""" Final Fitting. """
feffit_result = feffit(feffit_params, feffit_datasets)
```

```python
""" Plotting Fitted Results. """
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

spectra = feffit_result.datasets[0].data
fit = feffit_result.datasets[0].model

ax[0].plot(fit.r, fit.chir_mag, color="r", label="fitting")
ax[0].plot(spectra.r, spectra.chir_mag, lw=0, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
ax[0].axvline(1)
ax[0].axvline(6.5)
ax[0].set_xlim(0, 8)
ax[0].legend()

ax[1].plot(fit.k, fit.chi * fit.k ** 2, color="r", label="fitting")
ax[1].plot(spectra.k, spectra.chi * spectra.k ** 2, lw=1, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
ax[1].plot(fit.k, fit.kwin)
ax[1].legend()
```

```python
""" Feffit report """
print(feffit_report(feffit_result))
```
