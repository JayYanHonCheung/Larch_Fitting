```python
""" Cu Standard Fitting. """

""" 0. Convert file to ASCII"""

standard_path = "data\\Cu_foil_STD_merge_norm.nxs" # Normalised .nxs file

# Reading .nxs by h5py
file = h5py.File(standard_path, "r")
energy = np.array(file["processed"]["result"]["energy"])
data = np.array(file["processed"]["result"]["data"])

# To pandas then .dat
standard_path_dat = "data\\CuStandard.dat"
spectrum = np.array([energy, data])
df = pd.DataFrame(data=spectrum.T, columns=["Energy", "Intensity"])
df = df.set_index("Energy")
df.to_csv(standard_path_dat)
```

```python
""" 1. Reading Data """

# Initialisation
standard_path_dat = "data\\CuStandard.dat"
std_cu = read_ascii(standard_path_dat, labels="energy mu")
std_cu.groupname = "std_cu"

# Pre-processing - EXAFS fittings only need a valid k space
autobk(std_cu, pre_edge_kws={"nnorm": 0, "nvict": 0}, rbkg=1.0) # Need to have option to suppress normalisation in case loaded spectra is already normed
xftf(std_cu, kmin=3.0, kmax=10, dk=1.0, window="hanning", kweight=2.00)
xftr(std_cu, rmin=1.0, rmax=4.0, dr=1.0, window="hanning")
```

```python
""" Misc: Plotting normalisation results. """
dat = std_cu

plt.plot(dat.energy, dat.mu, label='mu')
plt.plot(dat.energy, dat.pre_edge, label='pre-edge line')
plt.plot(dat.energy, dat.post_edge, label='post-edge curve')
plt.plot(dat.energy, dat.norm, label='norm')
plt.legend()
plt.xlabel("Energy (eV)")
plt.ylabel("Normalised Absorption")
```

```python
""" Misc: Plotting k & R space """
dat = std_cu

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].plot(dat.energy, dat.norm)
ax[0].axhline(1, color="k", linestyle=":")
ax[1].plot(dat.k, dat.chi)
ax[2].plot(dat.r, dat.chir_mag)
```

```python
""" 2. Set up feffit. """

# Create feffit Parameter Group to hold fit parameters
_feffit_params = param_group(n1=param(12.0, vary=False),
                             s02=param(1.0, min=0, vary=True),
                             e0=param(0.1, min=-25, max=25, vary=True),
                             delr_Cu256=param(0.001, min=-0.5, max=0.5, vary=True),
                             sigma2_Cu256= param(0.008, min=0.0, vary=True))

# Define Fourier transform and fitting space
# Actual FT parameters for EXAFS fitting
_feffit_trans = feffit_transform(kmin=3.000, kmax=10.000, dk=1.0000, kw=2,
                                 window="hanning", fitspace='r', rmin=1.000, rmax=4.000)

# Make sure a dictionary for Feff Paths exists
try:
    npaths = len(_paths.keys())
except:
    _paths = {}
    
_paths['Cu256'] = feffpath("C:/Users/user/larch/feff/Cu1_K_Copper_cif11145/feff0001.dat",
                  label='Cu256', degen=1,
                  s02='n1 * s02',     e0='e0',
                  deltar='delr_Cu256', sigma2='sigma2_Cu256',
                  third='0', ei='0')

# Make dict of paths, sum them using a group of parameters
_ff2chi_paths  = {'Cu256': _paths['Cu256']}
if len(_ff2chi_paths) > 0:
    _pathsum = ff2chi(_ff2chi_paths, paramgroup=_feffit_params)
    
# Build feffit dataset, run feffit
_feffit_dataset = feffit_dataset(data=std_cu, transform=_feffit_trans, paths=_ff2chi_paths)
_feffit_result = feffit(_feffit_params, _feffit_dataset) # _feffit_dataset can be a list***

# Insert _feffit_result into self.feffit_history
if not hasattr(std_cu, 'feffit_history'): 
    std_cu.feffit_history = []    
_feffit_result.label = 'Fit %i' % (1+len(std_cu.feffit_history))
std_cu.feffit_history.insert(0, _feffit_result)
```

```python
""" Misc: Plotting fitted results"""

data_r = _feffit_dataset.data.r
data_chir_mag = _feffit_dataset.data.chir_mag

fit_r = _feffit_dataset.model.r
fit_chir_mag = _feffit_dataset.model.chir_mag

plt.plot(fit_r, fit_chir_mag, color="r", label="fit")
plt.plot(data_r, data_chir_mag, lw=0, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
plt.legend()
```
