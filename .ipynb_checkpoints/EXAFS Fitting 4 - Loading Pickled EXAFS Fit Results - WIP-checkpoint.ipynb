```python
"""
A demonstration in loading a "pickled" feffit result file. (meaning saved using the pickle package into a .pkl file)

"""
```

```python
from tools import io_results as io
import matplotlib.pyplot as plt
```

```python
""" 
Uses the result pickled in "EXAFS Fitting - Multipaths Fitting (Cu Foil Standard)" 

It saves all (meta)data of the EXAFS fit.
Below is a look into what category of data is stored.

"""

fpath = r"data/output/cu_foil_multipath/feffit_result.pkl"
result = io.load(fpath)
result
```

```python
result.datasets[0]
```

```python
""" Plotting Loaded Results. """
fig, ax = plt.subplots(1, 2, figsize=(20, 7))

spectra = result.datasets[0].data
fit = result.datasets[0].model
paths = result.datasets[0].paths

ax[0].plot(fit.r, fit.chir_mag, color="r", label="fitting")
ax[0].plot(spectra.r, spectra.chir_mag, lw=0, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
ax[0].axvline(1)
ax[0].axvline(6)
ax[0].set_xlim(0, 8)
ax[0].legend()

ax[1].plot(fit.k, fit.chi * fit.k ** 2, color="r", label="fitting")
ax[1].plot(spectra.k, spectra.chi * spectra.k ** 2, lw=1, marker='o', markersize=3, fillstyle='none', color="k", label="Raw Data")
ax[1].plot(fit.k, fit.kwin)
ax[1].legend()
```

```python
""" Values and Uncertainty of each parameter. """
s02 = result.paramgroup.s02

print(f"s02: {s02._val} +/- {s02.stderr}")

result.paramgroup.s02.__dict__
```
