```python
""" Misc: Plotting """
def plot(data_list, index):
    energy = data_list[index].energy
    mu = data_list[index].mu
    norm = data_list[index].norm
    chi = data_list[index].chi
    k = data_list[index].k
    kwin = data_list[index].kwin
    chir_mag = data_list[index].chir_mag
    r = data_list[index].r

    fig, ax = plt.subplots(1, 3, figsize=(15, 5))

    ax[0].plot(energy, norm)
    ax[1].plot(k, chi*k**2)
    ax[1].plot(k, kwin)
    ax[2].plot(r, chir_mag)
    
plot(data_list, 0)
```

```python
def fit_table(feff_folder):

    columns = ["scatter material", "spectra", "nlegs", "reff", "shell", "amp ratio", 
               "s02", "e0", "degen", "sigma2", "deltar", "fpath"]

    df = pd.DataFrame(columns=columns)
    feff_paths = finfo.get_path_info(feff_folder)
    
    df["fpath"] = [os.path.join(feff_folder, file) for file in feff_paths["file"]]
    df["amp ratio"] = feff_paths["amp ratio"]
    df["degen"] = feff_paths["deg"]
    df["nlegs"] = feff_paths["nlegs"]
    df["reff"] = feff_paths["r effective"]
    df["scatter material"] = finfo.get_path_formula(feff_folder)
    
    for column in columns:
        if df[column].isna().all():
            df[column] = column

    return df
```

```python
""" Bunch of helper functions 

Needs to packaged into a module.
"""

def setup_feffit_params(feff_paths):
    """ Helper function to setup feffit_params for feffit. """
    feffit_params = param_group()
    
    # Default parameters
    params_dict = {"s02": param(1.0, min=0.6, max=12.0, vary=True),
                   "degen": param(1, min=-1.0, vary=True),
                   "sigma2": param(0.008, min=0.001, max=0.015, vary=True),
                   "e0": param(0.1, min=-25.0, max=25.0, vary=True),
                   "deltar": param(0.1, max=0.5, vary=True),
                   "alpha": param(0.0, min=-0.1, max=0.1, vary=True)} 
    
    for key in params_dict:
        if key in feff_paths:
            if is_string_dtype(feff_paths[key]):
                for i, string in enumerate(feff_paths[key].unique()):
                    if not np.any([symbol in string for symbol in ["*", "/", "-", "+"]]):
                        setattr(feffit_params, string, params_dict[key])

    return feffit_params

def setup_feffit_paths(feff_paths):
    """ Helper function to setup path_dict for feffit. """
    paths_dict = {}
    
    for i, row in feff_paths.iterrows():
        label = f"{row['scatter material']}_{row['reff']}"
        paths_dict[label] = feffpath(row["fpath"],
                                     label=label,
                                     degen=row["degen"],
                                     e0=row["e0"],
                                     sigma2=row["sigma2"],
                                     s02=row["s02"],
                                     deltar=row["deltar"])
    
    return paths_dict

def form_shells(feff_paths):
    """ Categorise feff paths into shells labelled as a new column. 
        (WIP) For now, same reff is grouped into one shell
    """
    
    reff0 = None
    shell_num = 0
    shell = []
    for reff in feff_paths["reff"]:
        if reff == reff0:
            pass
        else:
            shell_num += 1
            
        reff0 = reff
        shell.append(shell_num)
    
    feff_paths["shell"] = shell
    
    return feff_paths

def groupby_path(feff_paths, keys):
    """ Group parameters based on scattering path/material. """
    
    for key in keys:
        feff_paths[key] = [f"{param}_{mat}" for param, mat in zip(feff_paths[key], feff_paths["scatter material"])]
    
    return feff_paths

def groupby_shell(feff_paths, keys):
    """ Group parameters based on scattering shell. """
    
    for key in keys:
        feff_paths[key] = [f"{param}_s{i}" for i, param in zip(feff_paths["shell"], feff_paths[key])]
    
    return feff_paths

def groupby_dataset(feff_paths, keys):
    """ Group parameters based on dataset. """
    
    for key in keys:
        feff_paths[key] = [f"{param}_d{i+1}" for param, dataset in zip(feff_paths[key], feff_paths["spectra"]) for i, _ in enumerate(feff_paths["spectra"].unique()) if _ == dataset]
    
    return feff_paths

def constant_param(feff_paths, key, param=None):
    """ Set constant params. """
     
    if param == None:
        param = key
        
    feff_paths[key] = param
    
    return feff_paths

def independent_params(feff_paths, keys):
    """ Set all parameters as independent. """
    
    for key in keys:
        feff_paths[key] = [f"{key}_{str(i+1)}" for i in feff_paths.index]
    
    return feff_paths

def _1st_shell_e0(feff_paths):
    """ Separate e0 into 1st shell and the rest. (Shelly Kelly's strategy) """
    
    e0 = []
    for i in feff_paths["shell"]:
        if i == 1:
            e0.append(f"e0_s{i}")
        else:
            e0.append(f"e0_s2")
    
    feff_paths["e0"] = e0
    
    return feff_paths

def _add_constant_params(feff_paths):
    """ Add feff path params for set constant parameters. """
    
    key_list = ["deg", "sig2", "e0", "s02"]
    for key in key_list:
        if is_numeric_dtype(feff_paths[key]):
            feff_paths[key + "_param"] = [f"{key}_{str(i + 1)}" for param in feff_paths[key] for i, uni_param in enumerate(feff_paths[key].unique()) if param == uni_param]
    
    return feff_paths
```
